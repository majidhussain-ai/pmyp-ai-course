A Regular Expression (RegEx) is a way to search for patterns in text. It can check if a word or pattern exists in a string or split the string into parts.

Using RegEx in Python

#### Why Use Regular Expressions in NLP?
Regular expressions (regex) are tools to find patterns in text.
They help clean, split, and search text data efficiently.
Regex is widely used in NLP to handle text data quickly, like removing special symbols, splitting sentences, or finding keywords.


#### Basic Components of Regular Expressions
Characters and character classes:
    [a-z]: Matches lowercase letters.
    \d: Matches any digit (0-9).
Quantifiers:
    *: Matches 0 or more times.
    +: Matches 1 or more times.
    {n}: Matches exactly n times.
Anchors:
    ^: Start of a string.
    $: End of a string.
Groups and capturing:
    ( … ): Groups patterns together.
Escaping special characters:
    \.: Matches a literal dot (.)

Metacharacters in Regular Expressions
Metacharacters are special symbols used in RegEx. Some common ones are:

\: Escapes special characters
[]: Matches characters in a group
^: Matches the beginning of a string
$: Matches the end of a string
.: Matches any character except newline
*, +, {}: Match multiple occurrences
|: OR (matches either pattern)
(): Groups parts of the expression


\ (Backslash) 
The backslash makes a character behave normally instead of as a metacharacter. 
For example, to search for a dot (.), you would write \. to avoid it being treated as a special character.

```python
import re

s = 'Hello.brother'

# Match any single character (wildcard)
match_any = re.search(r'.', s)
print("Any character match:", match_any)  # Matches 'H'

# Match a literal dot
match_dot = re.search(r'\.', s)
print("Literal dot match:", match_dot)  # Matches '.'

#Explanation:
#match_any finds the first character ('H') since . matches any character.
#match_dot finds the literal dot ('.') in the string.
```

[] (Square Brackets)
Square brackets define a character class, allowing you to match any character inside them.

```python
import re

# Pattern to match any vowel
pattern = r"[aeiou]"

# Sample texts
text1 = "Hello, welcome to the NLP course!"
text2 = "Artificial Intelligence is fascinating."

# Find all vowels in the first text
matches1 = re.findall(pattern, text1)
print("Matched vowels in text 1:", matches1)

# Find all vowels in the second text
matches2 = re.findall(pattern, text2)
print("Matched vowels in text 2:", matches2)

# Explanation:
#Two sample sentences (text1 and text2) are provided.
#The code uses re.findall(pattern, text) to find and print all vowels in both sentences.
#The output will show the matched vowels from each text separately.
```

^ (Caret)
The caret checks if a string starts with a specific character or characters. For example, ^g checks if a string begins with 'g'.

```python
import re

regex = r'^The'
strings = ['The quick brown fox', 'The lazy dog', 'A quick brown fox']
for string in strings:
    if re.match(regex, string):
        print(f'Matched: {string}')
    else:
        print(f'Not matched: {string}')

# Explanation:
# The regex pattern ^The checks if each string starts with "The."
# The code iterates through the strings list, using re.match() to determine if the pattern matches the beginning of each string.
# It prints whether each string matched or not, indicating whether it starts with "The."/#
```

#### \$ (Dollar)
The dollar checks if a string ends with a specific character or characters. For example, s$ checks if a string ends with 's'.

```python
# Sample sentence to search within
string = "The quick brown fox jumps over the lazy dog!"
pattern = r"dog!$"
object = re.search(pattern, string)
if object:
    print("Match found!")  # Output: Match found!
else:
    print("Match not found.")

# Explanation:
# The string now contains a complete sentence: "The quick brown fox jumps over the lazy dog!".
# The pattern r"dog!$" checks if the sentence ends with "dog!".
# The code searches for the specified pattern and prints "Match found!" if it's at the end, or "Match not found." if it isn't.
```

. (Dot)
The dot matches any single character except for a newline. For example, a.b matches 'acb', 'aXb', etc.

```python
import re

string = "The quick brown fox jumps over the lazy dog."
pattern = r"brown.fox"

match = re.search(pattern, string)
if match:
    print("Match found!")  # Output: Match found!
else:
    print("Match not found.")
```

| (Or)
The vertical bar works like an OR operator, matching either the pattern before or after it. For example, a|b matches 'a' or 'b'.

? (Question Mark)
The question mark means the preceding element can occur zero or one time. For example, ab?c matches 'ac' or 'abc' but not 'abbc'.

\* (Star)
The star matches zero or more occurrences of the preceding character. For example, ab*c matches 'ac', 'abc', or 'abbc'.

\+ (Plus)
The plus matches one or more occurrences of the preceding character. For example, ab+c matches 'abc' or 'abbc' but not 'ac'.

{m, n} (Braces)
Braces match a specific number of repetitions. For example, a{2,4} matches 'aa', 'aaa', or 'aaaa'.

( <regex> ) (Group)
Parentheses group sub-patterns together. For example, (a|b)cd matches 'acd' or 'bcd'.

Here’s a Python code example that demonstrates the use of various regular expression metacharacters, including |, ?, *, +, {m, n}, and () for grouping. Each example includes a string to search through and the corresponding regular expression pattern

```python
import re

# Sample sentence
sentence = "The lion roars, the leopard purrs, and monkeys swing from tree to tree."

# 1. Or Operator (|) - Finding either "lion" or "monkey"
print("1. Or Operator (|):")
pattern_or = r"lion|monkey"
match_or = re.findall(pattern_or, sentence)
print("   Found:", match_or)

# 2. Optional Character (?) - Find "leopard" with or without the 'd'
print("\n2. Optional Character (?):")
pattern_optional = r"leopar?d"  # 'd' is optional, so it will match "leopard" or "leopar"
match_optional = re.findall(pattern_optional, sentence)
print("   Found:", match_optional)

# 3. Zero or More (*) - Match "tre" followed by any number of 'e's
print("\n3. Zero or More (*):")
pattern_zero_or_more = r"tre*"
match_zero_or_more = re.findall(pattern_zero_or_more, sentence)
print("   Found:", match_zero_or_more)

# 4. One or More (+) - Match "purr" with one or more 'r's
print("\n4. One or More (+):")
pattern_one_or_more = r"pur+r"
match_one_or_more = re.findall(pattern_one_or_more, sentence)
print("   Found:", match_one_or_more)

# 5. Specific Repetitions ({m,n}) - Match "r" in "roar" repeated exactly 1 to 2 times
print("\n5. Specific Repetitions ({m,n}):")
pattern_specific_repetitions = r"roar{1,2}"
match_specific_repetitions = re.findall(pattern_specific_repetitions, sentence)
print("   Found:", match_specific_repetitions)

# 6. Grouping (()) - Match either "lion" or "leopard" followed by "purrs"
print("\n6. Grouping (()):")
pattern_grouping = r"(lion|leopard) purrs"
match_grouping = re.findall(pattern_grouping, sentence)
print("   Found:", match_grouping)
```

Overview of Special Sequences in Regular Expressions
Special sequences in regular expressions are tools that help you find patterns in strings based on their position or type of character. They don't match specific characters but instead indicate where a match should occur. Here’s a simple breakdown:

Why Use Special Sequences?
They make it easier to create flexible search patterns. Instead of writing complex regex, you can use these sequences to focus on the structure of the text, like whether a word starts or ends in a certain way, or whether it includes digits, spaces, or letters.

List of Special Sequences
\A: Checks if the string starts with something.

Use Case: Useful for validating that a string begins with a specific character or word.
\b: Marks the beginning or end of a word.

Use Case: Helpful for finding whole words, ensuring you don’t match parts of words.
\B: The opposite of \b; checks if the string does not start or end with a specific character.

Use Case: Useful when you want to avoid matching at word boundaries.
\d: Matches any digit (0-9).

Use Case: Great for finding numbers in text.
\D: Matches any non-digit character.

Use Case: Useful for finding letters or symbols, avoiding numbers.
\s: Matches any whitespace character (spaces, tabs).

Use Case: Helpful for parsing text where spaces matter.
\S: Matches any non-whitespace character.

Use Case: Good for finding continuous text without breaks.
\w: Matches any alphanumeric character (letters and digits).

Use Case: Useful for finding words or identifiers in code.
\W: Matches any non-alphanumeric character (punctuation, symbols).

Use Case: Great for filtering out letters and numbers, focusing on special characters.
\Z: Checks if the string ends with something.

Use Case: Useful for ensuring a string finishes with a specific word or character.

```python
import re

# Sample string for demonstration
sample_text = "Hello world! 123 cats and 456 dogs. @home $200.00"

# Using \A to check if the string starts with "Hello"
if re.match(r'\AHello', sample_text):
    print("Starts with 'Hello'")

# Using \b to find a word that starts with 'cat'
cat_matches = re.findall(r'\bcat\w*', sample_text)
print("Words starting with 'cat':", cat_matches)  # Output: ['cats']

# Using \B to check if 'at' is not at the start or end of a word
if re.search(r'\Bat', sample_text):
    print("Found 'at' not at word boundary")

# Using \d to find all digits
digits = re.findall(r'\d+', sample_text)
print("Digits found:", digits)  # Output: ['123', '456', '200', '00']

# Using \D to find all non-digit characters
non_digits = re.findall(r'\D+', sample_text)
print("Non-digit characters:", non_digits)

# Using \s to find all whitespace characters
whitespace = re.findall(r'\s+', sample_text)
print("Whitespace found:", whitespace)

# Using \S to find all non-whitespace characters
non_whitespace = re.findall(r'\S+', sample_text)
print("Non-whitespace characters:", non_whitespace)

# Using \w to find all alphanumeric characters
alphanumeric = re.findall(r'\w+', sample_text)
print("Alphanumeric characters:", alphanumeric)

# Using \W to find all non-alphanumeric characters
non_alphanumeric = re.findall(r'\W+', sample_text)
print("Non-alphanumeric characters:", non_alphanumeric)

# Using \Z to check if the string ends with "00"
if re.search(r'00\Z', sample_text):
    print("Ends with '00'")
```

Regular Expressions (RegEx) in Python
The re module in Python provides various functions to search and manipulate strings using regular expressions. Below are some key functions along with their descriptions and examples.

Key Functions of the re Module
#### re.findall(): Finds all occurrences of a pattern in a string and returns them as a list.
#### re.compile(): Compiles a regular expression pattern into a pattern object for more efficient matching.
#### re.split(): Splits a string based on occurrences of a pattern.
#### re.sub(): Replaces occurrences of a pattern with a specified string.
#### re.subn(): Similar to re.sub(), but returns a tuple containing the new string and the count of replacements.
#### re.escape(): Escapes special characters in a string so they can be treated as literal characters.
#### re.search(): Searches for the first occurrence of a pattern and returns a match object.

```python
import re

# Paragraph to analyze
paragraph = """
Python is a powerful programming language that is easy to learn. It supports a wide range of libraries, including those for data science, web development, and machine learning. 
Python was created in 1991 and is used by developers worldwide.
"""

# 1. re.findall()
# Find all whole words in the paragraph that start with 'p' or 'P'
pattern = r'\b[Pp]\w+'  # Pattern to find words starting with 'p' or 'P'
matches = re.findall(pattern, paragraph)  
print("1. Words starting with 'p' or 'P':", matches)  # Output: ['Python', 'powerful', 'programming', 'Python']

# 2. re.compile()
# Compile a pattern to find years (four-digit numbers)
year_pattern = re.compile(r'\b\d{4}\b')
years = year_pattern.findall(paragraph)  
print("2. Years found:", years)  # Output: ['1991']

# 3. re.split()
# Split the paragraph into sentences based on punctuation marks like periods, commas, or newlines
sentences = re.split(r'[.,\n]+', paragraph)
print("3. Sentences split by punctuation:", [s.split() for s in sentences if s.split()])  
# Output: ["Python is a powerful programming language that is easy to learn", 
#          "It supports a wide range of libraries, including those for data science, web development, and machine learning", 
#          "Python was created in 1991 and is used by developers worldwide"]

# 4. re.sub()
# Replace all occurrences of "Python" with "Java"
updated_paragraph = re.sub(r'Python', 'Java', paragraph)
print("4. Paragraph after replacement:", updated_paragraph)  
# Output: "Java is a powerful programming language that is easy to learn..."

# 5. re.subn()
# Replace "Python" with "Java" and count the number of replacements
updated_paragraph_count = re.subn(r'Python', 'Java', paragraph)
print("5. Replacement with count:", updated_paragraph_count)  
# Output: ("Java is a powerful programming language that is easy to learn...", 2)

# 6. re.escape()
# Escape any special characters in a phrase (if we want to search for the exact text in regex)
phrase = "Python is a powerful language!"
escaped_phrase = re.escape(phrase)
print("6. Escaped phrase:", escaped_phrase)  
# Output: 'Python\ is\ a\ powerful\ language\!'

# 7. re.search()
# Search for a specific pattern of a programming language followed by a year
search_pattern = r"([A-Za-z]+)\s(\d{4})"
match = re.search(search_pattern, paragraph)
if match:
    print("7. Full match:", match.group(0))  # Output: 'Python 1991'
    print("   Language:", match.group(1))   # Output: 'Python'
    print("   Year:", match.group(2))       # Output: '1991'
else:
    print("No match found.")
```

```python
import re

# Paragraph to work with
text = ("Data science is a growing field that combines programming, statistics, and domain expertise. "
        "Created in 2010, it has become essential for industries worldwide. Fields like finance, healthcare, "
        "and marketing use data science to make informed decisions.")

# 1. Find All Words Starting with "d" or "D"
# Pattern to find whole words starting with "d" or "D"
pattern1 = r'\b[Dd]\w*'  
words_with_d = re.findall(pattern1, text)
print("1. Words starting with 'd' or 'D':", words_with_d)

# 2. Identify Four-Digit Years
# Compiling a pattern to match four-digit numbers
pattern2 = re.compile(r'\b\d{4}\b')  
years = pattern2.findall(text)
print("2. Four-digit years found:", years)

# 3. Split the Paragraph into Phrases
# Splitting text by punctuation marks (periods and commas)
phrases = re.split(r'[.,]', text)
phrases = [phrase.strip() for phrase in phrases if phrase]  # Removing any extra spaces and empty items
print("3. Split into phrases:", phrases)

# 4. Replace a Specific Word
# Replacing "data" with "information" in the text
updated_text = re.sub(r'\bdata\b', 'information', text, flags=re.IGNORECASE)
print("4. Text with 'data' replaced by 'information':", updated_text)

# 5. Count Replacements of a Word
# Using re.subn to replace "science" with "analytics" and count replacements
updated_text_with_count, count = re.subn(r'\bscience\b', 'analytics', text, flags=re.IGNORECASE)
print("5. Text with 'science' replaced by 'analytics':", updated_text_with_count)
print("   Number of replacements:", count)

# 6. Escape Special Characters
# Escaping a phrase with special characters for regex
escaped_phrase = re.escape("data science is essential!")
print("6. Escaped phrase:", escaped_phrase)

# 7. Search for a Pattern of Field and Year
# Looking for a field (like "finance") followed by a four-digit year
pattern7 = r'\b(finance|healthcare|marketing)\b.*?(\d{4})'
match = re.search(pattern7, text, flags=re.IGNORECASE)
if match:
    print("7. Field and year found:", match.group(0))
    print("   Field:", match.group(1))
    print("   Year:", match.group(2))
else:
    print("7. No field and year found.")
```
