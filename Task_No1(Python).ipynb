## Task1. Keyword-Based Spam Checker
- This code takes user input, splits it into words, and checks for spam keywords.

```python
# Objective: Detect spam words in a user's message.

# Take input from the user and convert to lowercase for accurate matching
message = input("Enter your message: ").lower()

# Split the message into a list of words
words = message.split()

# Define the list of spam keywords
spam_keywords = ["win", "free", "prize"]

# Check if any spam word exists in the user's message
is_spam = False
for word in words:
    if word in spam_keywords:
        is_spam = True

# Output the result using an if condition
if is_spam:
    print("Spam Detected")
else:
    print("Safe Message")
```

## Task2. Simple Sentiment Detector
- This checks if the sentence has more positive words or negative words.

```python
# Objective: Analyze if a sentence is positive, negative, or neutral.

sentence = input("Enter a sentence: ").lower()
words = sentence.split()

# Define word lists for sentiment analysis
positive_words = ["good", "happy", "great", "excellent", "love"]
negative_words = ["bad", "sad", "terrible", "hate", "angry"]

pos_count = 0
neg_count = 0

# Count positive and negative words using loops
for word in words:
    if word in positive_words:
        pos_count += 1
    elif word in negative_words:
        neg_count += 1

# Determine the overall sentiment using comparison operators
if pos_count > neg_count:
    print("Sentiment: Positive")
elif neg_count > pos_count:
    print("Sentiment: Negative")
else:
    print("Sentiment: Neutral")
```

## Task5. Password Strength Checker
- This checks if the password is long enough and contains numbers or symbols.

```python
# Objective: Evaluate password strength based on length and characters.

password = input("Enter your password: ")

length_of_password = len(password)
has_number = False
has_symbol = False
symbols = "!@#$%^&*()-_+="

# Check the presence of numbers and symbols
for char in password:
    if char.isdigit():
        has_number = True
    elif char in symbols:
        has_symbol = True

# Evaluate password strength
if length_of_password >= 8 and has_number and has_symbol:
    print("Password Strength: Strong")
elif length_of_password >= 6 and (has_number or has_symbol):
    print("Password Strength: Medium")
else:
    print("Password Strength: Weak")
```

## Task6. Word Frequency Counter 
- This counts how many times each word is repeated in a sentence.

```python
# Objective: Count how many times each word appears in a sentence.

text = input("Enter a sentence: ").lower()

# Convert sentence into a list of words
words_list = text.split()

# Dictionary to store word counts
word_counts = {}

# Count frequencies
for word in words_list:
    if word in word_counts:
        word_counts[word] += 1
    else:
        word_counts[word] = 1

# Print the final counts
print("Word Frequencies:")
for word, count in word_counts.items():
    print(f"'{word}': {count} times")
```

## Task8. Data Classifier (Even/Odd Intelligence)
This separates a single list of numbers into two separate lists.

```python
# Objective: Separate a list of numbers into even and odd lists.

# Take multiple numbers in a list
numbers = [12, 5, 8, 19, 22, 3, 10, 7]

even_list = []
odd_list = []

# Classify data using modulo operator
for num in numbers:
    if num % 2 == 0:
        even_list.append(num)
    else:
        odd_list.append(num)

# Print both lists
print(f"Even List (Class A): {even_list}")
print(f"Odd List (Class B): {odd_list}")
```

## Task9. Smart Text Cleaner
 - This removes messy spaces and fixes the text automatically.

```python
# Objective: Clean extra spaces from a messy sentence to make it AI-ready.

messy_text = "   This   is   a    very    messy   sentence.   "

# 1. strip() removes starting and ending spaces
# 2. split() breaks it into words and automatically ignores extra spaces
words_list = messy_text.strip().split()

# 3. join() combines the words back with exactly one space between them
clean_text = " ".join(words_list)

print("Original:", repr(messy_text)) # repr() helps visualize the spaces
print("Cleaned text:", repr(clean_text))
```
