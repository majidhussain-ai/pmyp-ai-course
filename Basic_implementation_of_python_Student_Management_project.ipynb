```python
import json
import os

# Parent Class
class Person:
    def __init__(self, name, age):
        self.name = name  # Public variable
        self._age = age   # Protected variable

    def display_info(self):
        print(f"Name: {self.name}, Age: {self._age}")



```

```python
p = Person('majid', int(20))
p.display_info()
```

```python
# Child Class aur Inheritance
class Student(Person):
    # Class variable
    total_students = 0

    def __init__(self, name, age, student_id, course):
        super().__init__(name, age)
        self.__student_id = student_id  
        self.course = course
        Student.total_students += 1

    # Method overriding
    def display_info(self):
        print(f"ID: {self.__student_id}, Name: {self.name}, Age: {self._age}, Course: {self.course}")

    # Getter for private variable
    def get_student_id(self):
        return self.__student_id

    # Class method
    @classmethod
    def get_total_students(cls):
        return cls.total_students

    # Data convert karne ke liye helper methods
    def to_dict(self):
        return {
            "id": self.__student_id,
            "name": self.name,
            "age": self._age,
            "course": self.course
        }

    @classmethod
    def from_dict(cls, data):
        return cls(data['name'], data['age'], data['id'], data['course'])
```

```python

# Functions for CRUD & File Handling
class StudentManagementSystem:
    FILE_NAME = "students.json"

    def __init__(self):
        self.students = []
        self.load_data()

    def load_data(self):
        # Exception handling for file operations
        try:
            if os.path.exists(self.FILE_NAME):
                with open(self.FILE_NAME, 'r') as file:
                    data = json.load(file)
                    for item in data:
                        self.students.append(Student.from_dict(item))
                    Student.total_students = len(self.students)
        except Exception as e:
            print(f"Error loading data: {e}")

    def save_data(self):
        try:
            with open(self.FILE_NAME, 'w') as file:
                json.dump([student.to_dict() for student in self.students], file, indent=4)
        except Exception as e:
            print(f"Error saving data: {e}")

    # Input Validations
    def add_student(self):
        try:
            student_id = input("Enter Student ID: ").strip()
            if not student_id:
                raise ValueError("ID empty nahi ho sakti.")
            if any(s.get_student_id() == student_id for s in self.students):
                raise ValueError("Yeh Student ID pehle se mojood hai (Unique ID required).")

            name = input("Enter Name: ").strip()
            if not name:
                raise ValueError("Name empty nahi ho sakta.")

            age_input = input("Enter Age: ").strip()
            if not age_input.isdigit():
                raise ValueError("Age sirf numbers mein honi chahiye.")
            
            age = int(age_input)
            if age < 10 or age > 100:
                raise ValueError("Valid age (10-100) enter karein.")

            course = input("Enter Course: ").strip()
            if not course:
                raise ValueError("Course empty nahi ho sakta.")

            new_student = Student(name, age, student_id, course)
            self.students.append(new_student)
            self.save_data()
            print("\nStudent added successfully!")
            
        except ValueError as ve:
            print(f"\nValidation Error: {ve}")

    def view_students(self):
        if not self.students:
            print("\nKoi record mojood nahi hai.")
            return
        print("All Student Records")
        for student in self.students:
            student.display_info()
        print(f"Total Students: {Student.get_total_students()}")

    def search_student(self):
        student_id = input("Enter Student ID to search: ").strip()
        for student in self.students:
            if student.get_student_id() == student_id:
                print("Record Found ")
                student.display_info()
                
                # Name mangling demonstration
                # print(f"Accessing private var via name mangling: {student._Student__student_id}")
                return
        print("\nStudent not found.")

    def update_student(self):
        student_id = input("Enter Student ID to update: ").strip()
        for student in self.students:
            if student.get_student_id() == student_id:
                print("\nNew details enter karein (purani value rakhne ke liye khali chhor dein):")
                name = input(f"New Name ({student.name}): ").strip()
                age_input = input(f"New Age ({student._age}): ").strip()
                course = input(f"New Course ({student.course}): ").strip()

                if name: 
                    student.name = name
                if age_input:
                    try:
                        age = int(age_input)
                        if 10 <= age <= 100:
                            student._age = age
                        else:
                            print("Invalid age range. Age update nahi ki gayi.")
                    except ValueError:
                        print("Invalid input. Age update nahi ki gayi.")
                if course: 
                    student.course = course
                
                self.save_data()
                print("\nStudent updated successfully!")
                return
        print("\nStudent not found.")

    def delete_student(self):
        student_id = input("Enter Student ID to delete: ").strip()
        for student in self.students:
            if student.get_student_id() == student_id:
                self.students.remove(student)
                Student.total_students -= 1
                self.save_data()
                print("\nStudent deleted successfully!")
                return
        print("\nStudent not found.")
```

```python
#Menu Driven Program aur Conditional Statements
def main():
    sms = StudentManagementSystem()
    
    while True:
        print("Student Management System ")
        print("1. Add Student")
        print("2. View Students")
        print("3. Search Student")
        print("4. Update Student")
        print("5. Delete Student")
        print("6. Exit")
        
        choice = input("Enter your choice (1-6): ").strip()
        
        if choice == '1':
            sms.add_student()
        elif choice == '2':
            sms.view_students()
        elif choice == '3':
            sms.search_student()
        elif choice == '4':
            sms.update_student()
        elif choice == '5':
            sms.delete_student()
        elif choice == '6':
            print("System exit ho raha hai. Goodbye!")
            break
        else:
            print("Invalid choice! Please enter a number between 1 and 6.")

if __name__ == "__main__":
    main()
```
